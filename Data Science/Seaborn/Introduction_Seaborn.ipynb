```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
titanic = pd.read_csv('../titanic.csv')
```

countplot is used to show the counts of observations in each categorical bin using bars.
hue='pclass' is used to split data categories.

```python
plt.figure(figsize=(12, 8))
sns.set(font_scale=1.5)
sns.countplot(data=titanic, x='sex', hue='pclass')
plt.show()
```

# Categorical Plots

```python
plt.figure(figsize=(12, 8))
sns.set(font_scale=1.5)
sns.stripplot(data=titanic, x='sex', y='age', jitter=True, hue='pclass', dodge=True)
plt.show()
```

```python
plt.figure(figsize=(12, 8))
sns.set(font_scale=1.5)
sns.swarmplot(data=titanic, x='sex', y='age', hue='pclass', dodge=True)
plt.show()
```

```python
plt.figure(figsize=(12, 8))
sns.set(font_scale=1.5)
sns.barplot(data=titanic, x='pclass', y='age', hue='sex', dodge=True)
plt.show()
```

# Joint Plots / Regression Plots

kind='reg' is used to show the regression line.

```python
sns.set(font_scale=1.5)
sns.jointplot(data=titanic, x='age', y='fare', height=8, kind='reg')
plt.show()
```

```python
sns.set(font_scale=1.5)
sns.jointplot(data=titanic, x='age', y='fare', height=8, kind='kde')
plt.show()
```

```python
sns.set(font_scale=1.5)
sns.lmplot(data=titanic, x='age', y='fare', aspect=1, height=8, hue='sex')
plt.show()
```

```python
sns.set(font_scale=1.5)
sns.lmplot(data=titanic, x='age', y='fare', aspect=1, height=8, col='sex')
plt.show()
```

```python
sns.set(font_scale=1.5)
sns.lmplot(data=titanic, x='age', y='survived', aspect=1, height=8, col='sex', logistic=True)
plt.show()
```

# Matrix Plots / Heatmaps

```python
pd.crosstab(titanic.sex,  titanic.pclass)
```

```python
plt.figure(figsize=(12, 8))
sns.set(font_scale=1.5)
sns.heatmap(pd.crosstab(titanic.sex,  titanic.pclass), annot=True, fmt='d', cmap='Reds', vmax=None)
plt.show()
```

```python
pd.crosstab(titanic.sex, titanic.pclass, values=titanic.survived, aggfunc='mean')
```

```python

plt.figure(figsize=(12, 8))
sns.set(font_scale=1.5)
sns.heatmap(pd.crosstab(titanic.sex, titanic.pclass, values=titanic.survived, aggfunc='mean'), annot=True, cmap='Reds', vmax=None)
plt.show()
```

```python
numeric_cols = titanic.select_dtypes(include=[np.number])

numeric_cols.corr()      # correlation matrix
```

```python
plt.figure(figsize=(12, 8))
sns.set(font_scale=1.4)
sns.heatmap(numeric_cols.corr() , annot=True, cmap='Reds')
plt.show()
```
