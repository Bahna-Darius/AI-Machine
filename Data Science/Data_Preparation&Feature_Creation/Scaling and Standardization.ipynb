```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
titanic = pd.read_csv('../titanic.csv')
titanic.describe()
```

```python
titanic.fare.plot(figsize=(8,6))
titanic.age.plot(figsize=(8,6))
plt.show()
```

We can scale and standardize these features by calculating Z-score.
Calculate the Z-score of age and fare and store them in new columns called age_zscore and fare_zscore. 


```python
mean_age = titanic.age.mean()
mean_fare = titanic.fare.mean()
std_age = titanic.age.std()
std_fare = titanic.fare.std()

titanic['age_z'] = round((titanic.age - mean_age) / std_age, 2)
titanic['fare_z'] = round((titanic.fare - mean_fare) / std_fare, 2)

titanic.head()
```

```python
round(titanic.describe(), 2)
```

```python
titanic.fare_z.plot()
titanic.age_z.plot()
plt.show()
```
