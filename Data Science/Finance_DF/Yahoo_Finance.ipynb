# Normalizing Time Series to a Base Value (100)
### normalization helps us see how prices change over time, using the first price as a reference point, so it's easier to compare them and understand trends

```python
close.head()
```

```python
close.iloc[0]
```

ALL STOCKS START AT 100. It is a common practice to normalize the time series to a base value of 100. This allows us to compare the performance of the stocks starting at the same point. To normalize the time series to a base value of 100, we can divide each value in the time series by the first value in the time series and then multiply by 100.
For this always we start with first row of the data frame.

```python
norm = close.div(close.iloc[0]).mul(100)
norm
```

```python
norm.plot(figsize=(12, 8), fontsize=12)
plt.show()
```

The shift() method()

```python
aapl = close['AAPL'].copy().to_frame()  #to_frame() method is used to convert the series to dataframe
aapl
```

```python
aapl.shift(periods=2)
```

```python
aapl['lag1'] = aapl.shift(periods=1)
aapl
```

```python
aapl.AAPL.sub(aapl.lag1)    #sub() method is used to subtract the values of two series
```

```python
aapl['diff'] = aapl.AAPL.sub(aapl.lag1)
aapl
```

```python
aapl['pct_change'] = aapl.AAPL.div(aapl.lag1).sub(1).mul(100)
aapl
```

### The methods diff() and pct_change() are used to calculate the difference between consecutive values and the percentage change between consecutive values, respectively.

```python
aapl['diff'] = aapl.AAPL.diff(periods=1)    #diff() method is used to calculate the difference between consecutive values
aapl
```

```python
aapl['pct_change'] = aapl.AAPL.pct_change(periods=1).mul(100)    #pct_change() method is used to calculate the percentage change between consecutive values
```
