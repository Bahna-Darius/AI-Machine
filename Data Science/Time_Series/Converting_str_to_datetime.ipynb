# Converting strings to datetime objects with pd.to_datetime()

```python
import pandas as pd
```

```python
temp = pd.read_csv('../temp.csv')
temp.info()
```

### With the pd.to_datetime() function you can convert strings to datetime objects.

```python
data = pd.to_datetime(temp.datetime)
data.info()
```

Deleting the old column and setting the new datetime column as the index.

```python
temp.set_index(pd.to_datetime(temp.datetime)).drop(columns='datetime', axis=1)
```

Default format is year-month-day. You can change the format with the format parameter.
And 00:00:00 is added by default.

```python
pd.to_datetime('2019-01-01')
```

```python
pd.to_datetime('2019-01-01 12:32:00')
```

```python
pd.to_datetime('2019/01/01')
```

```python
pd.to_datetime('2019 01 03')
```

```python
pd.to_datetime('2019 May 03')
```

### You can also specify the format of the string.
### %Y: year, %m: month, %d: day, 
### %H: hour, %M: minute, %S: second

```python
pd.to_datetime('01-03-2019', format='%d-%m-%Y')
```

```python
pd.to_datetime(['01-03-2019', '02-03-2019'], format='%d-%m-%Y')
```

### If the string is not in the specified format, it will raise an error. With the errors parameter you can ignore the error.

```python
pd.to_datetime(['01-03-2019', '02-03-2019', 'Elephant'], format='%d-%m-%Y', errors='coerce')
```
