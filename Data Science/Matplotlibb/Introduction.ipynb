```python
import random

my_list = [random.normalvariate(0, 100) for i in range(1000)]

plt.hist(my_list, bins=100)  # bins = cate coloane
plt.show()
```

STYLE:


```python
plt.style.use('fivethirtyeight')    # Un stile tile mai curat
plt.hist(my_list, bins=100)
plt.show()
```

```python
plt.style.use('ggplot')   # Un stile care imita limbajul de programare R
plt.hist(my_list, bins=100)
plt.show()
```

```python
spped_bo = [random.normalvariate(50, 100) for c in range(1000)]
```

plt.vlines(sum(my_list)/ len(my_list), 0,50, color='red', linestyle='--', label='Mean New York') 

Acesta linie este un indicator pentru a arata media
Primul parametru este media, al doilea este de unde incepe linia (0,50 inseamna de la 0 la 50), culoarea, linestyle este stilul linie in cazul nostru este punctata, label este numele liniei

```python
plt.style.use('fivethirtyeight')
plt.figure(figsize=(12,6))
plt.hist(my_list, bins=100, label='New York Data',alpha=1, color='green')   # alpha = transparenta
plt.hist(spped_bo, bins=100, label='Boston Data', alpha=0.7, color='purple')
plt.title('Date De Test', fontsize=15)  # Titlul graficului
plt.xlabel('Speed') # Label pe axa x
plt.ylabel('Ceva Inaltime') # Label pe axa y
plt.vlines(sum(my_list)/ len(my_list), 0,50, color='red', linestyle='--', label='Mean New York')  
plt.vlines(sum(spped_bo)/ len(spped_bo), 0,50, color='blue', linestyle='--', label='Mean Boston')
# plt.axis((0, 100, 0, 50))
# plt.grid(True)  # Default acest style are grid
# plt.xticks(range(1,400, 2))
# plt.yticks(range(1,400, 2))
plt.legend(loc='upper right', fontsize=12)  # loc = pozitia legendei
plt.show()
```
