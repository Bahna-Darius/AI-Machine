```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
titanic = pd.read_csv('titanic.csv')
```

```python
xtickss = [x for x in range(0, 901, 50)]
ytickss = [ x for x in range(0, 81, 5)]
```

# Graph with plot

```python
titanic.age.plot(figsize=(8, 6), fontsize = 13, c="red", linestyle="-",
                 xlim=(0,900), ylim=(0,80), xticks=xtickss, yticks=ytickss,
                 rot=45
                 )
plt.title("Titanic - Age", fontsize=15)
plt.legend(fontsize=15)
plt.xlabel("Passenger No", fontsize=13)
plt.ylabel("Age", fontsize=13)
plt.grid()
plt.show()
```

# Histogram

```python
titanic.age.hist(
    figsize= (8,6), bins=80, xlabelsize=15, ylabelsize=15
)
plt.show()
```
