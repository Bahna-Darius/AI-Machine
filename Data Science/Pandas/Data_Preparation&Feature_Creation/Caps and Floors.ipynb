```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
titanic = pd.read_csv('../../Data/titanic.csv')
titanic
```

Here we have very large values for 'fare' which can affect our model because very large or very small data can affect model performance. So we need to normalize the data.

```python
titanic.fare.plot(figsize=(8, 6))
plt.show()
```

```python
titanic.describe()
```

```python
fare_cap = 250
```

```python
titanic.loc[titanic.fare > fare_cap, 'fare'] = fare_cap     # Overwriting max values with cap value because we can't have fare more than 250
```

```python
fare_floor = 5
titanic.loc[titanic.fare < fare_floor, 'fare'] = fare_floor     # Overwriting min values with floor value because we can't have 0 fare
```
