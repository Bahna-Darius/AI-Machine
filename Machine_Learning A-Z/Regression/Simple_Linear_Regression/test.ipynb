```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('Salary_Data.csv')
X = data.iloc[:, :-1].values
y = data.iloc[:, -1].values
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

```python
from sklearn.linear_model import LinearRegression

regressor = LinearRegression()
regressor.fit(X_train,y_train)
```

```python
predictions = regressor.predict(X_test)
predictions
```

```python
from sklearn.metrics import accuracy_score

scor = accuracy_score(y_test, predictions)
scor
```

# Visualising the Train data

```python
plt.style.use('fast')
plt.figure(figsize=(8,7))
plt.scatter(X_train, y_train, color='purple', label='Train data')
plt.plot(X_train, regressor.predict(X_train), color = 'red', label='Simple liniar reg.')
plt.title('Salary VS XP', fontsize=15)
plt.xlabel('XP', fontsize=15)
plt.ylabel('Salary', fontsize=15)
plt.legend(fontsize=11)

plt.show()
```

# Visualising the Test Data

```python
plt.style.use('seaborn-v0_8-darkgrid')
plt.scatter(X_test, y_test, color='red', label='Test Data')
plt.plot(X_test, regressor.predict(X_test), color='yellow', label='Predict Data')
plt.title('Data Predict Model', fontsize=12)
plt.xlabel('XP', fontsize=15)
plt.ylabel('Salary', fontsize=15)
plt.legend(fontsize=11)
plt.show()
```
